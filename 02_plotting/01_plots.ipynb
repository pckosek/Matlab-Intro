### Plotting in MATLAB

Plots are pretty easy in MATLAB. In my opinion, the power of plots and the ability they provide to visualize your data is one of the major motivations to use this program. Let's take a look a plotting. 

First, we'll generate a simple cosine wave, and then plot the data.

```matlab
x = cos( 2*pi*[0:9]/ 10)
plot( x )
```

So, this is nice, but we can make it nicer. 

The above visualization (which we can't zoom in on) could benefit from markers. Here let's the specify a black plot, and connected 'plus' markers, 

```matlab
plot( x, 'k+-' )
```

MATLAB can plot two graphs at once, which is very, very nice.  The first way is to use the `hold` method.

```matlab
x = cos( 2*pi*1*[0:999]/1000 );
y = cos( 2*pi*2*[0:999]/1000 );
plot( x ); 
hold on;
plot( y );
hold off;
```

The second way to plot multiple data points is to feed your data as multiple columns. This works well - as long as you have the same number of data points in your data sets.

```matlab
x = [ [1:10]', [2:2:20]' ]
plot( x )
```

Again, make sure your data is in columns when you attempt to make multiple plots in this matter. Otherwise, your plot will wind up looking like this:

```matlab
x = [ 1:10; 2:2:20 ]
plot( x )
```

Something we haven't pointed out is that the x axis is the index in each of the above plots. This is an easy enough fix, but adding a first parameter to the `plot` command

```matlab
t = [0:999]/1000;
y = cos( 2*pi*3* t);
plot( t, y )
```

Plots of unequal number of data points can be compared using the `hold` command.

```matlab
t1 = [0:9]/10;
t2 = [0:49]/50;
t3 = [0:999]/1000;
plot( t1, cos(2*pi*3*t1), '+' );
hold on;
plot( t2, cos(2*pi*3*t2), 'o' );
plot( t3, cos(2*pi*3*t3) );
hold off;
```

Just for funsies, let's add some labels

```matlab
t1 = [0:9]/10;
t2 = [0:49]/50;
t3 = [0:999]/1000;
plot( t1, cos(2*pi*3*t1), '+' );
hold on;
plot( t2, cos(2*pi*3*t2), 'o' );
plot( t3, cos(2*pi*3*t3) );
xlabel('time');
ylabel('amplitude')
hold off;
```
